```python
#Import the required libarary and import the KNeighbourClassifier for applying the KNN algorithm
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
from sklearn.neighbors import KNeighborsClassifier
```

```python
#Read the csv file 
#Use the seperation command to divide the columns(option)
df1 = pd.read_csv("Cities.csv", sep=',', engine='python')
df1.head()
```

```python
#plot a scatter graph for showing the varisity of the latitudes and longitudes
plt.scatter(df1.Latitude,df1.Longitude,c="Blue")
plt.rcParams['figure.figsize'] = (10,10)
plt.grid(True)
```

```python
#Divide the training data and the test data
#Keep the majority for training data - 85%
train_data = int(len(df1)*0.85)
test_data = len(df1) - train_data
print('Training data size = ',train_data,'\t','Testing data size = ',test_data)
```

```python
#Dividing the data set along the size of the training data and testing data
cities_train = df1[0:train_data]
cities_test = df1[train_data:len(df1)]
```

```python
#Markdown the features for which you have to analyse and predict
#You can use the latitude longitude and Temperature (remember use the integeral column only)
#Decide with how many neighbours you wanna analyse the given features k =8(here)
features = ['Latitude','Longitude']
k = 8

classifier = KNeighborsClassifier(k)
classifier.fit(cities_train[features], cities_train["Category"])
predictions = classifier.predict(cities_test[features])
```

```python
#Creating an array that only consist the Category of the test_data
category_test = np.array(cities_test.Category)

#comparing the predicted results and actual results and analysing how many are correct
correct = 0
for i in range(0,len(predictions)):
    if predictions[i] == category_test[i]:
        correct+=1

print('Total correct Predictions = ',correct)
```
